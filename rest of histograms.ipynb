```python
import pandas as pd
import os
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
'''
  Reads the 150MB+ file named 'all_gazepoints_max_invalid_per_subject10%.txt'
  into a Pandas DataFrame.

  Inputs
    None
  Returns
    pd.DF containing data from file
'''
def read_txt_into_df(txt_filename):
#     df = pd.read_csv(txt_filename, delim_whitespace=True, comment='/')
    df = pd.read_csv(txt_filename, delim_whitespace=True, comment='/', names = ["category", "start_frame", "end_frame"])
    print(df.info())
    return df
```

```python
def format_hand_code(fpath):
    list_df = []
    for root, dirs, files in os.walk(fpath):
        for name in files:
            df = read_txt_into_df(root + '/' + name)
            list_df.append((df, name))
            
    return list_df
```
